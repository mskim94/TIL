# 데이터 윤리

## 26.1 데이터 윤리란 무엇인가?
데이터의 사용과 함께 데이터의 오용도 발생한다. 

e.g. 

2016년 대선에서, 캠브리지 아날리티카라른 회사가 페이스북의 데이터를 부적절하게 접근하여 이를 정치적 광고 타겟팅에 사용했다.

2018년에는 우버에서 실험한 자율중행 자동차가 보행자를 치여 숨지게 하는 일이 있었다(당시 '안전 담당 운전자'가 운전석에 있었으나 주의를 기울이지 않고 있었다.)

그리고 범죄자들의 재범률을 예측하기 위한 알고리즘이 사용되고, 이에 따라 형량이 부과되고 있다. 과연 이것이 판사들이 형량을 결정하는 것보다 더 공정하다고 할 수 있을까?


## 26.2 아니, 진짜로, 데이터 윤리가 뭔데?

'윤리란 무엇인가?' **윤리(Ethics)** 란 '옳고 그른' 행동을 판단하기 위한 틀 정도로 귀결된다. 그렇다면 '데이터 윤리'란 데이터 관련된 행동의 옳고 글므을 판단하기 위한 틀이라고 볼 수 있다.


## 26.3 데이터 윤리에 대해 신경 써야 할까?

뉴스 탐색 알고리즘에 가해진 아주 작은 변화가 어떤 글이 수백만 명이 읽는 글이 되거나 아무도 읽지 않는 글이 되는 차이를 가져다 줄 수 있다. 

이 처럼 어떤 상황에서도 여러분의 업무가 세상에 미치는 영향에 대해 생각해 봐야 한다. 특히, 업무의 영향력이 클수록 이러한 부분을 더욱 많이 고민해봐야 한다.


## 26.4 나쁜 데이터 제품 만들기

'데이터 윤리' 이슈는 '나쁜 제품'을 만들어서 생긴 결과이다.

가령, MS에서 트윗을 받으면 이를 흉내 내는 테이(Tay)라는 이름의 챗봇을 배포했는데, 이를 통해 인터넷 사용자들이 금세 테이가 온갖 불쾌한 트윗을 하게 만들 수 있었다. MS 안의 어느 누구도 '인종차별' 챗봇을 배포하는 것의 윤리성에 대해 고민해 본 사람은 없었을 것이다. 봇이 어떻게 오용될 수 있을지 충분히 검토해 보지 못한 채 그저 챗봇을 만들었을 것이다.

## 26.5 정확도와 공정함의 균형을 유지하기

'정확도'와 '공정함' 사이에 트레이드오프 관계가 있을 수 있다는 것, 이러한 트레이드오프에는 '올바른' 해답이 항상 존재하지 않을 수 있다는 점이다.

## 26.6 협력
억압적인 국가의 정부 관료들이 드디어 해당 국가의 시민들이 데이텀에 가입해도 좋다고 승인했다. 하지만 해당 국가의  사용자들이 딥러닝에 대해 논의를 해서는 안 된다고 주장했다. 게다가, 딥러닝에 대한 정보를 알아내려고 시도하는 사용자들의 이름을 보고해 주길 원한다.

해당 국가의 데이터 과학자들에게 이런 식으로 주제가 제한된 데이텀에 접근할 수 있도록 제공하는 것이 나을까? 혹은 제안된 제약이 너무 끔찍하므로 아예 서비스를 제공하지 않는 것이 나을까?

## 26.7 해석 가능성
어떤 상황에는 모델의 성능이 비록 더 나쁘더라도 결과를 해석할 수 있는 모델을 더 선호할 수도 있다. 다른 상황에서는, 가장 성능이 좋은 모델을 원할 수도 있을 것이다. 또 어떤 경우에는 해석이 가능하고 성능도 좋은 모델이 존재하지 않을 수도 있다.

## 26.8 추천

유뷰브에서 영상을 볼 때, 유튜브 측에서 다음에 볼 영상을 추천해 준다. 유튜브는 광고를 토앻 돈을 벌기 때문에 여러분이 볼 확률이 높은 영상을 추천해 주어 더 많은 광고를 보여 주고 싶어 할 것이다. 하지만 사람들은 음로 이론 관련 영상을 보는 것을 좋아하기에 관련 영상이 추천되는 경향이 있음이 드러났다.

구글은 불쾌하거나 폄하하는 내용을 포함하는 자동완성 검색어를 직접 걸러 내고 있다.

## 26.9 편향된 데이터

21.6절 '단어 벡터'에서 우리는 문서의 말뭉치를 사용해서 단어의 벡터 임베딩을 학습했다. 이 벡터는 '분포적 유사도'를 유지하도록 만들어졌다. 즉, 비슷한 맥락에 등장하는 단어들은 비슷한 벡터로 표현된다. 이는 학습 데이터 안의 편향이 단어 벡터에도 반영된다는 것을 의미한다.

e.g. 

만약 문서가 R 사용자는 윤리적으로 타락한 사람들이고 파이썬 사용자는 미덕의 귀감이라는 내용을 담고 있다면, 모델은 높은 확률로 파이썬과 R에 대해 이러한 관계를 학습할 것이다.

만약 대다수의 뉴스에서 소프트웨어 엔지니어를 다루는 기사가 남성 소프트웨어 엔지니어에 대한 것이었다면 학습된 '소프트웨어' 벡터는 '여성' 단어 벡터 보다 '남성'단어 벡터에 가깝게 위치할 수 있다.

이렇게 학습된 벡터를 사용하는 애플리케이션도 비슷하게 편향된 유상성을 지닐수 있다. 이런 경우에 특정 '편향'을 제거하는 기법들이 다양하게 존재하지만, 모든 편향을 제거하는 것은 아마 불가능할 것이다. 하지만 그러한 방법이 있음을 알아둘 필요가 있다.

26.4절 나쁜 데이터 제품 만들기의 '사진' 예제와 같이, 모델을 대표성이 없는 데이터를 사용하여 학습한다면 해당 모데르이 실제 성능이 안 좋을 가능성이 매우 높으며, 난처한 방식으로 동작할 수도 있다.

## 26.10 데이터 보호
원하는 대로 데이터를 사용해도 무방한 사용자 계약을 맺은 상황이다.

누군가는 광고주에게 데이터를 제공하는 것이 옳지 않다고 생각하고, 누군가는 학계 안의 사람들은 데이터를 책임지고 보호할 수 없을 것이라고 걱정한다. 다른 누군가는 회사가 정치와 관련되어서는 안된다고 생각하고, 마지막 사람은 경찰은 신뢰할 수 없으면 법 집행기관과 협업하면 결백한 사람에게 해를 끼칠 것이라고 주장한다.

## 26.11 요약
