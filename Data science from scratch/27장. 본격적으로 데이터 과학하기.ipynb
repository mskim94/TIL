IPython을 마스터하면 삶이 훨씬 편해진다.

## 27.3 밑바닥부터 시작하지 않는 방법

'밑바닥부터' 무언가를 만드는 것은 그것이 어떻게 동작하는지 배우는 데 큰 도움이 된다. 하지만 (처음부터 성능을 염두에 두고 만든 것이 아니라면) 그런 경우에는 보통 성능이 뛰어나지 않다. 게다가 사용성이 나쁘거나, 빠른 프로토타이핑에 부적합거나, 오류 핸들링(error handling)이 잘 되지 않을 수 있다.

실전에서는 기초부터 조금 더 탄탄하게 구현된 라이브러리를 사용하는 것이 좋을 수 있다. 

### 27.3.1 Numpy 

Numpy는 'Numeric Python'의 약자로, '진짜' 과학적인 계산을 할 때 사용되는 다양한 도구를 제공한다. 가령 리스트로 만든 벡터보다 좋은 성능을 보이는 배열(array)이라든지, 리스트의 리스트 보다 좋은 성능을 보이는 행렬(matrix)을 비롯해, 그들을 다룰 수 있는 각종 함수를 제공한다.

Numpy는 다른 많은 라이브러리에서도 사용되므로, 특히 알아주면 좋다.

### 27.3.2 Pandas

Pandas는 파이썬으로 데이터를 다룰 때 사용하면 편한 추가적인 자료 구조를 제공한다. 그중에서도 특히 DataFrame 이 중요하다. DataFramedms 24장 '데이터 베이스와 SQL'에서 구축한 NotQuiteABase Table과 개념적으로 유사하지만 기능이나 성능면에서 우월하다.

파이썬으로 데이터셋을 합치고 나누고, 뭉치고, 수정하는 등의 작업을 한다면 pandas는 매우 유용한 도구이다.

### 27.3.3 scikit-learn

scikit-learn은 파이썬 머신러닝 라이브러리 중에 가장 인기가 있을 것이다. 이책에서 만든 대부분의 모델이 구현되어 있을 뿐만 아니라, 그 외에도 다양한 모델을 제공한다. 실제 문제를 풀 때는 의사결정나무를 '밑바닥부터에서부터' 만들기보다는 scikit-learn을 쓰게 될 것이다. 실제 문제를 다룰 때는, 최적화 알고리즘을 처음부터 작성하기보다는 scikit-learn의 성능을 믿고 사용하게 될 것이다.

scikit-learn의 문서에는 상당히 많은 예시가 포함되어 있다. (머신러닝이 어떤 상황에서 유용한지 까지 상세히 설명되어 있다.)


### 27.3.4 시각화

matplotlib으로 생성한 그래프들은 깔끔하고 가능적이기는 하지만 세련되었다고는 할 수 없다. 데이터 시각화에 조금 더 관심이 있다면 몇 가지 옵션이 더 있다.

seaborn이라는 라이브러리도 있는데, 이 라이브러리는 무엇보다 matplotlib을 더욱 매력적으로 만들어 준다.

웹을 위한 인터랙티브 시각화를 만들고 싶다면 'Data Driven Document(데이터 기반 문서)'라고 불리는 자바스크립ㅌ 라이브러리 D3.js를 써보자.

자바스크립트에 대해 아는 것이 거의 없더라도 D3 갤러리에 있는 예시들을 이용하면 내 데이터에 맞게 사용할 수 있다.

D3에 전혀 관심이 없더라도, 갤러리를 한번쯤 훑어보는 것만으로 시각화에 대해 많은 것을 배울 수 있다.

마지막으로 Bokeh는 D3의 기능을 파이썬으로 가져온 것이다.


### 27.3.5 R

R을 배우지 않아도 아무런 문제가 되지 않지만, 많은 데이터 과학 프로젝트가 R을 사용하기 때문에 익숙해지는 건 의미가 있다.

그래야 R에 기반한 블로그 포스트, 예시, 또는 코드를 이해할 수 있기 때문이다. 


### 27.3.6. 딥러닝

딥러닝을 하지 않고도 데이터 과학자가 될 수 있지만, 트렌디한 데이터 과학자가 되지는 못할 것이다.

가장 인기 있는 파이썬 딥러닝 프레임워크로는 구글의 텐서플로(TensorFlow)와 페이스북의 파이토치(PyTorch)가 있다. 인터넷에서 최악부터 최고까지 다 양한 스펙트럼의 튜토리얼을 찾을 수 있다.

이 중에서 텐서플로가 조금 더 오래되고 널리 사용되고 있는데, 내 생각에는 파이토치가 훨씬 쉽고 특히 초보자에게 친숙하게 느껴질 것이다.
