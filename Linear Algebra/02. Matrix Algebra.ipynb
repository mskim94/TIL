 # 2.1 Matrix Operations


## 용어 정리

- Matrix Notation: 행렬 표기법
- e.g. 
- Row $i$ $\begin{bmatrix}
{a}_{11} & ... & {a}_{1j} & ... & {a}_{1n}\\
: & \, & : &  & :\\
{a}_{i1}& ... & {a}_{ij} & ... & {a}_{in}\\
: & \, & : & \, & : \\
{a}_{m1} & ... & {a}_{mj} & ... & {a}_{mn}\\
\end{bmatrix}
$ 각 column은 ${a}_1$ ${a}_j$, ${a}_n$ 으로 표기

- The Transpose of a Matrix: row와 column을 바꾸어 놓은 Matrix, Matrix A의 Transpose Matrix는 ${A}^T$

## Summary

### Theorem 1.
A, B, and C 를 같은 사이즈의 matrix 라고 하고 r,s 를 scalar라고 하자.
- a. A + B = B + A
- b. (A + B) + C = A + (B + C)
- c. A + 0 = A
- d. r(A + B) = rA + rB
- e. (r + s)A = rA + sA
- f. r(sA) = (rs)A
를 만족한다.

#### Matrix Multiplication
A가 m x n  matrix이고 B가 column ${b}_1,...,{b}_p$를 가지고 n x p matrix이면 AB의 product는 m x p이고 AB의 column은 $A{b}_1,...,A{b}_p$이다. 
- $AB$ = A[${b}_1 {b}_2 ... {b}_p$] = [$A{b}_1,...,A{b}_p$]
- $({AB})_{ij}$ = ${a}_{i1}{b}_{1j} + {a}_{i2}{b}_{2j}+ ... + {a}_{in}{b}_{nj}$


➡️ column을 이용하여 product의 결과의 **특정 column만**을 구할 수 있다.

### Theorem 2.
Matrix A를 m x n 이라 하면 B, C는 sums 과 products로 정의되어 갖은 사이즈를 갖는다.
- a. A(BC) = AB(C)
- b. A(B + C) = AB + AC
- c. (B + C)A  = BA + CA
- d. r(AB) = (rA)B = A(rB) for any scalar r
- e. ${I}_mA = A = A{I}_n$

**warnings**

- In general, AB $\neq BA$
- If AB = BA, then **not** true in general that B = C
- If AB = 0, then **not** true in general A = 0 or B = 0

### Theorem 3.
- a. ${({A}^T)}^T$ = A
- b. $({A + B})^T$ = ${A}^T$ + ${B}^T$
- c. For and scalar r, ${(rA)}^T$ = r${A}^T$
- d. (${AB}^T$) = ${B}^T{A}^T$

 # 2.2 The Inverse of a Matrix


## 용어 정리
- Invertible Matrix:
CA = I and AC = I 를 만족하는 n x n의 C가 있다면, n x n Matrix A는 invertible하다.

- nonsingular matrix : invertible
- singular matrix: **not** invertible

## Summary
#### Invertible Matrix
- square matrix(정방 행렬)일때만 정의가 가능하며 invertible matrix는 unique하다. 
- Matrix가 inverse Matrix 가진다면 ➡️ invertible 하다 ➡️ nonsingular 하다


### Theorem 4.
$A = \begin{bmatrix}
a & b \\
c & d \\
\end{bmatrix}$ 이고 if $ad-bc \neq 0$ 이면 A는 invertible 하고 ${A}^{-1}$ = $\frac{1}{ad-bc}\begin{bmatrix}
d & -b \\
-c & a \\
\end{bmatrix}$ 

$detA = ad - bc$

### Theorem 5.
A가 invertible n x n Matrix이고 $\mathbb{R}^n$ 공간안의 각각의 b가 있이면, $Ax = b$의 unique solution $x = {A}^{-1}b$이다.

### Theorem 6.
- A가 invertible matrix이라면 ${A}^{-1}$이 invertible이고 ${({A}^{-1})}^{-1} = A$ 이다.
- $({AB})^{-1} = {B}^{-1}{A}^{-1}$
- ${({A}^T)}^{-1} = {({A}^{-1})}^{T}$

#### Elementary Matrices
: Elementary Matrix는 identity matrix에 single elementary row operation을 진행하고 얻은 Matrix이다.

- e.g.
- ${E}_1 = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
-4 & 0 & 1 \\
\end{bmatrix}$: replacement,  ${E}_2 = \begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$: interchange,  ${E}_3 = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 5 \\
\end{bmatrix}$: scaling  

$A = \begin{bmatrix}
{a}_{11} & {a}_{12} & {a}_{13} & {a}_{14}\\
{a}_{21} & {a}_{22} & {a}_{23} & {a}_{24}\\
{a}_{31} & {a}_{32} & {a}_{33} & {a}_{34}\\
\end{bmatrix}$ 
- ${E}_1A = \begin{bmatrix}
{a}_{11} & {a}_{12} & {a}_{13} & {a}_{14}\\
{a}_{21} & {a}_{22} & {a}_{23} & {a}_{24}\\
{a}_{31} - 4{a}_{11} & {a}_{32} - 4{a}_{11} & {a}_{33} - 4{a}_{13}& {a}_{34} - 4{a}_{14}\\
\end{bmatrix}$ - replacement 됨 
- ${E}_2A = \begin{bmatrix}
{a}_{21} & {a}_{22} & {a}_{23} & {a}_{24}\\
{a}_{11} & {a}_{12} & {a}_{13} & {a}_{14}\\
{a}_{31} & {a}_{32} & {a}_{33} & {a}_{34}\\
\end{bmatrix}$ - interchange 됨
- ${E}_3A = \begin{bmatrix}
{a}_{11} & {a}_{12} & {a}_{13} & {a}_{14}\\
{a}_{21} & {a}_{22} & {a}_{23} & {a}_{24}\\
5{a}_{31} & 5{a}_{32} & 5{a}_{33} & 5{a}_{34}\\
\end{bmatrix}$ - scaling 됨 

- 만약 elementary row operation이 m x n matrix A에 수행되면 그 결과 matrix는 EA로 표현될 수 있다. 
- m x m matrix E는 Identity matrix인 ${I}_m$에 same row operation을 수행하여 얻을 수 있다.
- 각 elementary matrix E는 invertible 하다. E의 역행렬은 E를 다시 I로 변환하는 같은 유형의 기본 행렬이다.

e.g.
- ${E} = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$

- ${E}_1 = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
-4 & 0 & 1 \\
\end{bmatrix}$ replacement : 3rd row **-** 4 x 1st row

- ${E}_2 = \begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$ interchange : 1st row **$\leftrightarrow$** 2nd row

- ${E}_3 = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 5 \\
\end{bmatrix}$ scaling : 5 **x** 3rd row

- ${E}_1^{-1} = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
4 & 0 & 1 \\
\end{bmatrix}$ replacement : 3rd row **+** 4 x 1st row (+ $\leftrightarrow$ -)

- ${E}_2^{-1} = \begin{bmatrix}
0 & 1 & 0 \\
1 & 0 & 0 \\
0 & 0 & 1 \\
\end{bmatrix}$ interchange : 1st row **$\leftrightarrow$** 2nd row

- ${E}_3^{-1} = \begin{bmatrix}
1 & 0 & 0 \\
0 & 1 & 0 \\
0 & 0 & 1/5 \\
\end{bmatrix}$ scaling : 3rd row **/** 5


### Theorem 7.
- n x n matrix A가 invertible 하다. $\Longleftrightarrow$ A는 ${I}_n$에 row equivalent 하다. (필요충분조건)
- 어느 elementary row operation 순서로 A를 ${I}_n$로 reduce 하면 같은 elementary row operation을 통해 ${A}^{-1}$로 변형할 수 있다.
    - $A$ $\to$ elementary row operation $\to$ $I$ 
    - $I$ $\to$ elementary row operation $\to$ ${A}^{-1}$     
    
#### Algorithm for Finding ${A}^{-1}$
$[A I] ~ [I C]$, $C = {A}^{-1}$
- e.g.
- $A = \begin{bmatrix}
0 & 1 & 2 \\
1 & 0 & 3 \\
4 & -3 & 8 \\
\end{bmatrix}$

- $\begin{bmatrix}
0 & 1 & 2 & | & 1 & 0 & 0\\
1 & 0 & 3 & | & 0 & 1 & 0\\
4 & -3 & 8 & | & 0 & 0 & 1\\
\end{bmatrix}$: 1. ~ $\begin{bmatrix}
1 & 0 & 3 & | & 0 & 1 & 0\\
0 & 1 & 2 & | & 1 & 0 & 0\\
4 & -3 & 8 & | & 0 & 0 & 1\\
\end{bmatrix}$: 2. ~ $\begin{bmatrix}
1 & 0 & 3 & | & 0 & 1 & 0\\
0 & 1 & 2 & | & 1 & 0 & 0\\
0 & -3 & -4 & | & 0 & -4 & 1\\
\end{bmatrix}$: 3.
- ~ $\begin{bmatrix}
1 & 0 & 3 & | & 0 & 1 & 0\\
0 & 1 & 2 & | & 1 & 0 & 0\\
0 & 0 & 2 & | & 3 & -4 & 1\\
\end{bmatrix}$: 4. ~ $\begin{bmatrix}
1 & 0 & 3 & | & 0 & 1 & 0\\
0 & 1 & 2 & | & 1 & 0 & 0\\
0 & 0 & 1 & | & 3/2 & -2 & 1/2\\
\end{bmatrix}$: 5. ~ $\begin{bmatrix}
1 & 0 & 3 & | & -9/2 & 7 & -3/2\\
0 & 1 & 2 & | & -2 & 4 & -1\\
0 & -3 & -4 & | & 3/2 & -2 & 1/2\\
\end{bmatrix}$: 6.

- ${A}^{-1} = \begin{bmatrix}
-9/2 & 7 & -3/2 \\
-2 & 4 & -1 \\
3/2 & -2 & 1/2 \\
\end{bmatrix}$

- 1. ~ 4. : forward phase, 5. ~ 6. : backward phase

#### Another View of Matrix Inversion
- $[A I]$ ~ \[$I {A}^{-1}$\]
- \[$A$ ${e}_1 {e}_2 ... {e}_n$\]
- Ax = ${e}_1$, Ax = ${e}_2$,..., Ax = ${e}_n$
- \[$I$ ${A}^{-1}{e}_1 {A}^{-1}{e}_2 ... {A}^{-1}{e}_n$\] ➡️ ${A}^{-1}$의 특정 column만을 알 수 있다.
