 # 2.1 Matrix Operations


## 용어 정리

- Matrix Notation: 행렬 표기법
- e.g. 
- Row $i$ $\begin{bmatrix}
{a}_{11} & \dots & {a}_{1j} & \dots & {a}_{1n}\\
\vdots & \, & \vdots &  & \vdots\\
{a}_{i1}& \dots & {a}_{ij} & \dots & {a}_{in}\\
\vdots & \, & \vdots & \, & \vdots \\
{a}_{m1} & \dots & {a}_{mj} & \dots & {a}_{mn}\\
\end{bmatrix}
$ 각 column은 ${a}_1$ ${a}_j$, ${a}_n$ 으로 표기

- The Transpose of a Matrix: row와 column을 바꾸어 놓은 Matrix, Matrix A의 Transpose Matrix는 ${A}^T$

## Summary

### Theorem 1.
A, B, and C 를 같은 사이즈의 matrix 라고 하고 r,s 를 scalar라고 하자.
- a. A + B = B + A
- b. (A + B) + C = A + (B + C)
- c. A + 0 = A
- d. r(A + B) = rA + rB
- e. (r + s)A = rA + sA
- f. r(sA) = (rs)A
를 만족한다.

#### Matrix Multiplication
A가 m x n  matrix이고 B가 column ${b}_1,...,{b}_p$를 가지고 n x p matrix이면 AB의 product는 m x p이고 AB의 column은 $A{b}_1,...,A{b}_p$이다. 
- $AB$ = A[${b}_1 {b}_2 ... {b}_p$] = [$A{b}_1,...,A{b}_p$]
- $({AB})_{ij}$ = ${a}_{i1}{b}_{1j} + {a}_{i2}{b}_{2j}+ ... + {a}_{in}{b}_{nj}$


➡️ column을 이용하여 product의 결과의 **특정 column만**을 구할 수 있다.

### Theorem 2.
Matrix A를 m x n 이라 하면 B, C는 sums 과 products로 정의되어 갖은 사이즈를 갖는다.
- a. A(BC) = AB(C)
- b. A(B + C) = AB + AC
- c. (B + C)A  = BA + CA
- d. r(AB) = (rA)B = A(rB) for any scalar r
- e. ${I}_mA = A = A{I}_n$

**warnings**

- In general, AB $\neq BA$
- If AB = BA, then **not** true in general that B = C
- If AB = 0, then **not** true in general A = 0 or B = 0

### Theorem 3.
- a. ${({A}^T)}^T$ = A
- b. $({A + B})^T$ = ${A}^T$ + ${B}^T$
- c. For and scalar r, ${(rA)}^T$ = r${A}^T$
- d. (${AB}^T$) = ${B}^T{A}^T$

 # 2.2 The Inverse of a Matrix


## 용어 정리
- Invertible Matrix:
CA = I and AC = I 를 만족하는 n x n의 C가 있다면, n x n Matrix A는 invertible하다.

- nonsingular matrix : invertible
- singular matrix: **not** invertible

## Summary
#### Invertible Matrix
- square matrix(정방 행렬)일때만 정의가 가능하며 invertible matrix는 unique하다. 
- Matrix가 inverse Matrix 가진다면 ➡️ invertible 하다 ➡️ nonsingular 하다


### Theorem 4.
$A = \begin{bmatrix}
a & b \\
c & d \\
\end{bmatrix}$ 이고 if $ad-bc \neq 0$ 이면 A는 invertible 하고 ${A}^{-1}$ = $\frac{1}{ad-bc}\begin{bmatrix}
d & -b \\
-c & a \\
\end{bmatrix}$ 

$detA = ad - bc$

### Theorem 5.
A가 invertible n x n Matrix이고 $\mathbb{R}^n$ 공간안의 각각의 b가 있이면, $Ax = b$의 unique solution $x = {A}^{-1}b$이다.

### Theorem 6.
- A가 invertible matrix이라면 ${A}^{-1}$이 invertible이고 ${({A}^{-1})}^{-1} = A$ 이다.
- $({AB})^{-1} = {B}^{-1}{A}^{-1}$
- ${({A}^T)}^{-1} = {({A}^{-1})}^{T}$

### Theorem 7.
- n x n matrix A가 invertible 하다. $\Longleftrightarrow$ A는 ${I}_n$에 row equivalent 하다. (필요충분조건)
- 어느 elementary row operation 순서로 A를 ${I}_n$로 reduce 하면 같은 elementary row operation을 통해 ${A}^{-1}$로 변형할 수 있다.
    - $A$ $\to$ elementary row operation $\to$ $I$ 
    - $I$ $\to$ elementary row operation $\to$ ${A}^{-1}$

$[A I] ~ [I C]$, $C = {A}^{-1}$

- ~ $\begin{bmatrix}

# 2.3 Characterizations of Invertible Matrices

## 용어 정리


## Summary
- Let A be a square n x n matrix.
    - a. **A** is an **invertible** matrix
    - b. There is an **n x n** matrix **C** such that **CA = I**
    - c. The equation **Ax = 0** has only the **trivial** solution
    - d. A has **n pivot** positions
    - e. A is **row equivalent** to the identity n x n matrix
    - f. There is an **n x n** matrix **D** such that AD = I
    - g. The equation Ax = b has at least one solution for each b in $\mathbb{R}^n$
    - h. The columns of **A span $\mathbb{R}^n$**
    - i. The linear transformation X $\mapsto$ Ax maps $\mathbb{R}^n$ onto $\mathbb{R}^n$
    - j. The columns of A form a ilnearly independent set
    - k. The linear transformation x $\mapsto$ Ax is one-to-one
    - l. ${A}^{T}$ is an invertible matrix
- 위 문장들은 모두 참이거나 모두 거짓이다.
- +e). [A I] ~ \[$I {A}^{-1}$\]: ${A}^{-1}$이 존재하면, Identity matrix로 row reduction 가능

### Theorem 9.
Let T: $\mathbb{R}^n$ $\to$ $\mathbb{R}^n$ be a linear transformation이고 A를 T에 standard matrix 라 할때, T is invertible $\Longleftrightarrow$ A is invertible matrix

# 2.4 Partitioned Matrices

## 용어 정리
- Partitioned Matirx: Matrix를 특정 row, column으로 나누어 매트릭스 안에 matrix block으로 만들어 놓은 matrix
- conformable: 적합, A, B가 연산이 가능하게 Partitioned 되어 있다면 그 연산에 conformable하다.

## Summary
#### Multiplication of Partitioned Matrices:
Matrix A의 column을 3 colums, 2 columns로 나누었다면  Matrix B 의 row는 3 rows, 2 rows로 나눠야  Multiplication이 가능하다.
- The partitions of A and B are conformable for block multiplication

### Theorem 10. Column-Row Expansion of AB
if A is mxn and B is n x p, then AB = \[${col}_1(A) {col}_2(A) ... {col}_n(A)$\]$\begin{bmatrix}
{row}_1(B) \\
{row}_2(B) \\
\dots \\
{row}_n(B) \\
\end{bmatrix}$ = ${col}_1(A){row}_1(B) + ... + {col}_n(A){row}_n(B)$ = \[$A{b}_1 ... A{b}_p$\]

#### Inverses of Partitioned Matrices
block upper triangular 형태의 Matrix A = $\begin{bmatrix}
{A}_{11} & {A}_{12} \\
0 & {A}_{22} \\
\end{bmatrix}$ (${A}_{11}$는 p x p, ${A}_{22}$는 q x q), ${A}^{-1} = \begin{bmatrix}
{A}_{11}^{-1} & {-A}_{11}^{-1}{A}_{12}{A}_{22}^{-1} \\
0 & {A}_{22}^{-1}
\end{bmatrix}$

# 2.5 Matrix Factorization

## 용어 정리
- Factorization: 
Matrix를 matrix의 곱으로 표현한 것
- LU Factorization:
Matrix를 lower triangle matrix와 Echelon form의 product로 표현 한 것

## Summary

#### LU Factorization (Interchange를 하지 않는 경우)
1. Matrix A 를 row replacement operation을 통해 reduce하여 echelon form U로 만든다.
2. 위의 Echelon form을 만드는 과정에서의 Pivot position을 뽑아 scaling을 통해 L을 만든다.

- e.g.
    - $A = \begin{bmatrix}
            2 & 4 & -1 & 5 & -2 \\
            -4 & -5 & 3 & -8 & 1 \\
            2 & -5 & -4 & 1 & 8 \\
            -6 & 0 & 7 & -3 & 1 \\
            \end{bmatrix}$ ~ $\begin{bmatrix}
            2 & 4 & -1 & 5 & -2 \\
            0 & 3 & 1 & 2 & -3 \\
            0 & -9 & -3 & -4 & 10 \\
            0 & 12 & 4 & 12 & -5 \\
            \end{bmatrix}$ ~ $\begin{bmatrix}
            2 & 4 & -1 & 5 & -2 \\
            0 & 3 & 1 & 2 & -3 \\
            0 & 0 & 0 & 2 & 1 \\
            0 & 0 & 0 & 4 & 7 \\
            \end{bmatrix}$ ~ $\begin{bmatrix}
            2 & 4 & -1 & 5 & -2 \\
            0 & 3 & 1 & 2 & -3 \\
            0 & 0 & 0 & 2 & 1 \\
            0 & 0 & 0 & 0 & 5 \\
            \end{bmatrix}$
     - pivot position: $\begin{bmatrix}
                         2 &  &  &    \\
                         -4 & 3 &  &   \\
                         2 & -9 & 2 &    \\
                         -6 & 12 & 4 & 5   \\
                         \end{bmatrix}$ 1st column / 2, 2nd column / 3, 3rd column / 2, 4th column /5
    - $L = \begin{bmatrix}
            1 & 0 & 0 & 0 \\
            -2 & 1 & 0 & 0 \\
            1 & -3 & 1 & 0 \\
            -3 & 4 & 2 & 1 \\
            \end{bmatrix}$
            
#### General Case (interchange가 필요한 경우)
 - **PA = LU**
 - Matrix A를 reduced row operation을 통해 U로 만든후 Pivot position을 뽑아 Identity matrix에 적용하는데 interchange가 된 부분이 digonal의 pivot entry이면 제외 한다.


#### LU Factorization 
- 1. Ax = b 를 LUx = b로 표현
- 2. 새로운 벡터 y를 Ux = y로 정의하고, LUx = b를 Ly = b로 표현
- 3. Ly = b에서 y를 구한다.
- 4. 구한 y를 Ux = y에 대입하여 최종적으로 x를 구한다.
- 두 번 풀어서 더 고생스러울 것 같지만, 실제로는 복잡한 A를 푸는 것 보다 쉬울 때가 있다.

# 2.6 Subspaces of $\mathbb{R}^n$

## 용어 정리
- Subspace: 부분 공간

## Summary

#### Subsapce of $\mathbb{R}^n$
Any set $H$ in $\mathbb{R}^n$ that has three properties:
- a. The zero vector in $H$
- b. For each $u$ and $v$ in $H$, the sum $u + v$ is in H
- c. For each $u$ in $H$ and scalar $c$. the vector $cu$ is in $H$
- $H$를 subspace 라고 한다.
- +) subspace는 zero vector를 포함하고 sum, scalar multiplication에 닫혀있어야 한다.

#### Column Space
Matrix A의 column space는 A의 column의 모든 linear combinations의 $Col$ $A$의 set이다.
- $A$ = \[${a}_1 ... {a}_n$\] $\mathbb{R}^m$
- $Col$ $A$ = Span{${a}_1,...,{a}_n$}
- Subspace of $\mathbb{R}^m$

Is b in $Col$ $A$? = Is $Ax = b$ consistent?
- +) matrix A의 column vector들로 만들수 있는 subspace

#### Null Space
Matrix A의 **null space**는 homogeneous equation $Ax = 0$의 모든 solutions의 $Nul A$ set 이다.

Is u in $Nul$ $A$? = Is $Au = 0$?
- +) Ax = 0을 만족시키는 모든 x의 집합
- +) Null space의 basis를 구하려면 Ax = 0 즉, homogeneous equation을 풀면된다. $\to$ augmented matrix(마지막 열에 0을 추가하여 echelon form으로 바꾸고 basis를 찾으면 된다.)

### Theorem 11.
The **null space** of an m x n Matrix A는 $\mathbb{R}^n$의 subspace이다.
- **null space** is defined implicitly
- **column space** is defined explicitly


#### Basis for a Subspace
$\mathbb{R}^n$의 subspace인 $H$에 basis는 H를 span하는 $H$의 linearly independent set이다.

#### Standard Basis for $\mathbb{R}^n$
$\mathbb{R}^n$을 span하는 Identity column vector의 set
- {${e}_1,...,{e}_n$}
- $\mathbb{R}^n$은 $\mathbb{R}^n$ 자체의 subspace이다.

#### Find a basis for the column space of Matrix
**row operations**는 matrix의 column들 사이에 linear dependence relations에 영향을 주지 않는다.
주의) Matrix A를 Echelon form or reduced echelon form으로 만드는 이유는 pivot position을 찾기 위함이지 A의 column space의 basis는 변형 전의 column들이다.

### Theorem 12.
matrix A의 pivot columns는 A의 column space에 basis를 형성한다.

#### row operations do not affect linear dependence relation among the columns of the matrix

# 2.7 Dimension and Rank

## 용어 정리
- coordinate: 좌표

## Summary
#### Coordinates and Coordinate Vector
subspace $H$에 대한 basis를 $B$ = {${b}_1,...,{b}_n$} set이라고 가정하면, $H$안의 each x에 대해, basis $B$에 관련된 x의 coordinates는 **weights** ${c}_1,...,{c}_p$로 $x = {c}_1{b}_1 + ... + {c}_p{b}_p$로 나타낼 수 있고,
${[x]}_{B} = \begin{bmatrix}
              {c}_{1} \\
              \vdots \\
              {c}_p \\
              \end{bmatrix}$ 은 x의 coordinate vector(relative to $B$) or x의 $B$-coordinate vector 라고 한다.
              
#### Dimension
nonzero subspace $H$의 dimension은 $dim H$, $H$에 basis안의 vector의 갯수이다. 
zero subspace {0}의 dimension은 zero로 정의될 수 있다.
- basis는 유일하지 않을 수 있다. but 개수는 동일하다. $\to$ dim은 같다.
- A plane through 0 in $\mathbb{R}^3$is two-dimensional
- A line through 0 in $\mathbb{R}^3$is one-dimensional

#### Rank
Matrix A의 rank는 rank $A$로 표기하며, A의 column space의 dimension이다.(column space의 Basis의 vector의 개수)

- e.g.
$A = \begin{bmatrix}
      2 & 5 & -3 & -4 & 8 \\
      4 & 7 & -4 & -3 & 9 \\
      6 & 9 & -5 & -2 & 4 \\
      0 & -9 & 6 & 5 & -6 \\
      \end{bmatrix}$ ~ $\begin{bmatrix}
      2 & 5 & -3 & -4 & 8 \\
      0 & -3 & 2 & 5 & -7 \\
      0 & 0 & 0 & 4 & -6 \\
      0 & 0 & 0 & 0 & 0 \\
      \end{bmatrix}$ 
- pivot column $\to$ A의 Basis: 3개 $\to$ rank $A$, dim $Nul A$ = 2 (free variable의 개수)
- (basic variable은 free variable로 표현이 가능하고 free variable의 개수에 따라 다른 basic variable의 개수가 정해진다.)

### Theorem 13. The Rank Theorem
If Matrix A has **n columns**, then rank$A$ + dim $Nul A$ = n

### Theorem 14. The Basis Theorem
- $H$를 $\mathbb{R}^n$의 p-dimensional의 subspace라고 하면, $H$안의 $p$ elements의 linearly independent set은 automatically $H$의 basis이다.
- Also, $H$를 span하는 $H$의 $p$ elements의 set은 $H$에 대한 basis이다.

### Theorem. The Invertible Matrix Theorem(2.3 continued)
Let A be an invertible n x n matrix. Then the following statements are equivalent.
- m. The columns of A form a basis of $\mathbb{R}^n$
- o. $Col$ $A$ = $\mathbb{R}^n$
- p. rank $A$ = n
- q. dim $Nul$ $A$ = 0
- r. $Nul$ $A$ = {0}
